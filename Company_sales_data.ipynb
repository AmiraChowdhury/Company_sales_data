```python
# import pandas library
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('company_sales_data.csv')
df.head()
```

```python
"""Exercise 1: Read Total profit of all months and show it using a line plot
Total profit data provided for each month. Generated line plot must include the following properties: –
● X label name = Month Number
● Y label name = Total profit"""

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(df['month_number'], df['total_profit'], marker='o')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title('Total Profit Over the Months')
plt.grid()
plt.show()
```

```python
"""Exercise 2: Get total profit of all months and show line plot with the following Style properties
Generated line plot must include following Style properties: –
● Line Style dotted and Line-color should be red
● Show legend at the lower right location.
● X label name = Month Number
● Y label name = Sold units number
● Add a circle marker.
● Line marker color as read
● Line width should be 3"""

plt.figure(figsize=(10, 5))
plt.plot(df['month_number'], df['total_profit'], linestyle=':', color='red', linewidth=3, marker='o', markerfacecolor='red', label='Total Profit')
plt.xlabel('Month Number')
plt.ylabel('Sold Units Number')
plt.legend(loc='lower right')
plt.title('Total Profit Over the Months (Styled)')
plt.grid()
plt.show()
```

```python
"""Exercise 3: Read all product sales data and show it using a multiline plot
Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline
for each product)."""

plt.figure(figsize=(10, 5))
plt.plot(df['month_number'], df['facecream'], label='Face Cream')
plt.plot(df['month_number'], df['facewash'], label='Face Wash')
plt.plot(df['month_number'], df['toothpaste'], label='Toothpaste')
plt.plot(df['month_number'], df['bathingsoap'], label='Bathing Soap')
plt.plot(df['month_number'], df['shampoo'], label='Shampoo')
plt.plot(df['month_number'], df['moisturizer'], label='Moisturizer')

plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.title('Product Sales Over the Months')
plt.legend()
plt.grid()
plt.show()
```

```python
"""Exercise 4: Read toothpaste sales data of each month and show it using a scatter plot. Also, add a grid in
the plot. gridline style should “–“."""

plt.figure(figsize=(10, 5))
plt.scatter(df['month_number'], df['toothpaste'])
plt.grid(linestyle='--')
plt.xlabel('Month Number')
plt.ylabel('Toothpaste Sales')
plt.title('Toothpaste Sales Scatter Plot')
plt.show()
```

```python
"""Exercise 5: Read face cream and facewash product sales data and show it using the bar chart"""

plt.figure(figsize=(10, 5))
labels = df['month_number']
x = range(len(labels))

bar_width = 0.35
plt.bar(x, df['facecream'], width=bar_width, label='Face Cream')
plt.bar([p + bar_width for p in x], df['facewash'], width=bar_width, label='Face Wash')

plt.xlabel('Months')
plt.ylabel('Units Sold')
plt.title('Face Cream and Face Wash Sales')
plt.xticks([p + bar_width / 2 for p in x], labels)
plt.legend()
plt.show()
```

```python
"""Exercise 6: Read sales data of bathing soap of all months and show it using a bar chart. Save this plot to
your hard disk"""

plt.figure(figsize=(10, 5))
plt.bar(df['month_number'], df['bathingsoap'])
plt.xlabel('Months')
plt.ylabel('Units Sold')
plt.title('Bathing Soap Sales')
plt.savefig('bathing_soap_sales.png')  # Save the plot
plt.show()
```

```python
"""Exercise 7: Read the total profit of each month and show it using the histogram to see the most common
profit ranges"""

plt.figure(figsize=(10, 5))
plt.hist(df['total_profit'], bins=5, edgecolor='black')
plt.xlabel('Profit Ranges')
plt.ylabel('Frequency')
plt.title('Histogram of Total Profit')
plt.show()
```

```python
"""Exercise 8: Calculate total sale data for last year for each product and show it using a Pie chart"""
total_sales = {
    'Face Cream': df['facecream'].sum(),
    'Face Wash': df['facewash'].sum(),
    'Toothpaste': df['toothpaste'].sum(),
    'Bathing Soap': df['bathingsoap'].sum(),
    'Shampoo': df['shampoo'].sum(),
    'Moisturizer': df['moisturizer'].sum(),
}

plt.figure(figsize=(8, 8))
plt.pie(total_sales.values(), labels=total_sales.keys(), autopct='%1.1f%%', startangle=140)
plt.title('Total Sales Data for Last Year')
plt.axis('equal')
plt.show()
```

```python
"""Exercise 9: Read Bathing soap facewash of all months and display it using the Subplot"""
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.bar(df['month_number'], df['bathingsoap'])
plt.title('Bathing Soap Sales')
plt.xlabel('Months')
plt.ylabel('Units Sold')

plt.subplot(1, 2, 2)
plt.bar(df['month_number'], df['facewash'])
plt.title('Face Wash Sales')
plt.xlabel('Months')
plt.ylabel('Units Sold')

plt.tight_layout()
plt.show()
```

```python
"""Exercise 10: Read all product sales data and show it using the stack plot"""
plt.figure(figsize=(10, 5))
plt.stackplot(df['month_number'],
               df['facecream'],
               df['facewash'],
               df['toothpaste'],
               df['bathingsoap'],
               df['shampoo'],
               df['moisturizer'],
               labels=['Face Cream', 'Face Wash', 'Toothpaste', 'Bathing Soap', 'Shampoo', 'Moisturizer'])

plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.title('Stack Plot of All Product Sales')
plt.legend(loc='upper left')
plt.show()
```
